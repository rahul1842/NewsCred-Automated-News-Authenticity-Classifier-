NLP

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
ps=WordNetLemmatizer()
```

```python
stopwords=stopwords.words('english')
```

```python
nltk.download('wordnet')
```

```python
def clean_row(row):
    
    row=row.lower()
    row=re.sub('[^a-zA-Z]',' ',row)
    
    token = row.split()
    
    news=[ps.lemmatize(word) for word in token if not word in stopwords]
    cleanned_news=' '.join(news)
    return cleanned_news
```

```python
dataset['title']
```

```python
dataset['title']=dataset['title'].apply(lambda x :clean_row(x))
```

```python
dataset['title']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features =115, lowercase=False , ngram_range=(1,2))
```

```python
X = dataset.iloc[:90,0]
Y = dataset.iloc[:90,1]
```

```python
X.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data,train_label,test_label=train_test_split(X,Y,test_size=0.2,random_state=0) 
```

```python
train_data
```

```python
# Ensure both training and testing data have the same shape
vec_train_data = vectorizer.fit_transform(train_data)
vec_test_data = vectorizer.transform(test_data)

# Convert to dense arrays for DataFrame compatibility
vec_train_data_array = vec_train_data.toarray()
vec_test_data_array = vec_test_data.toarray()

# Check the shapes to ensure they match
print(vec_train_data_array.shape)
print(vec_test_data_array.shape)

# Create DataFrames
train_df = pd.DataFrame(vec_train_data_array, columns=vectorizer.get_feature_names_out())
test_df = pd.DataFrame(vec_test_data_array, columns=vectorizer.get_feature_names_out())

# Check shapes of DataFrames
print(f"Training data shape: {train_df.shape}")
print(f"Testing data shape: {test_df.shape}")
```

Model

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
```

```python
y_pred = clf.predict(test_df)
```

```python

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Separate the text for real and fake news
fake_text = dataset[dataset['Label'] == 1]['title'].str.cat(sep=' ')
real_text = dataset[dataset['Label'] == 0]['title'].str.cat(sep=' ')

# Generate word clouds
fake_wc = WordCloud(width=800, height=400, background_color='black').generate(fake_text)
real_wc = WordCloud(width=800, height=400, background_color='white').generate(real_text)

# Plot side-by-side
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.imshow(fake_wc, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Fake News', fontsize=14)

plt.subplot(1, 2, 2)
plt.imshow(real_wc, interpolation='bilinear')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np

# Sum TF-IDF scores across all documents for each term
tfidf_scores = vec_train_data.sum(axis=0).A1
terms = vectorizer.get_feature_names_out()

# Map terms to their TF-IDF scores
term_scores = dict(zip(terms, tfidf_scores))

# Generate word cloud
wc = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(term_scores)

# Display word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Top Words in News Corpus Based on TF-IDF', fontsize=16)
plt.show()
```
